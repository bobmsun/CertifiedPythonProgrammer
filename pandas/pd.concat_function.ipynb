This notebook is used to demonstrate the use of `pd.concat` function and the parameter of this function:

* `aixs` parameter
* `join` parameter

```python
import pandas as pd
```

```python
df_1 = pd.DataFrame({
    'Col_A': [1, 2, 3, 4],
    'Col_B': [5, 6, 7, 8]
})

df_2 = pd.DataFrame({
    'Col_A': [9, 10, 11, 12],
    'A_Col_C': [13, 14, 15, 16]
})

df_2 = df_2.drop(index=[0, 2])
```

```python
pd.concat([df_1, df_2], axis='index', join='inner')
```

```python
pd.concat([df_1, df_2], axis='index', join='outer')
```

```python
# 以下对比是用来解释上面 warning 的意思，和 sort parameter 的意义
display(pd.concat([df_1, df_2], axis='index', join='outer', sort=False))
display(pd.concat([df_1, df_2], axis='index', join='outer', sort=True))
```

```python
pd.concat([df_1, df_2], axis='columns', join='inner')
```

```python
pd.concat([df_1, df_2], axis='columns', join='outer')
```

This notebook is to demonstrate:
* What can cause "SettingWithCopyWarning"?
* How to avoid "SettingWithCopyWarning"?
* Use .loc and .iloc for assignment?

```python
import pandas as pd
```

```python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'age': [42, 52, 36, 24, 73],
            'female': [0, 1, 1, 0, 1],
            'preTestScore': [47, 24, 31, 28, 33],
            'postTestScore': [25, 94, 57, 62, 70]
           }
df = pd.DataFrame(raw_data)
df
```

```python
df_2 = df[df['age'] > 50]
df_2['preTestScore'] = 1
display(df_2)
display(df)
```

```python
df_3 = df.query("age > 50")
df_3['preTestScore'] = 2
display(df_3)
display(df)
```

```python
df_4 = df.loc[df['age'] > 50]
df_4['preTestScore'] = 3
display(df_4)
display(df)
```

```python
df_4_2 = df.iloc[[1, 4]]
df_4_2['preTestScore'] = 33
display(df_4_2)
display(df)
```

```python
df[df['age'] > 50]['preTestScore'] = 5     # df[df['age'] > 50, 'preTestScore']  会出 error，不能这么用
df
```

```python
df.loc[df['age'] > 50, 'preTestScore'] = 4
df
```

```python
# df.iloc[[1, 3], 'preTestScore'] = 6 会出错，iloc 里不能用 'preTestScore' column name

df.iloc[[1, 3], df.columns.get_indexer(['preTestScore'])] = 6    
df   
```

```python
# 用 .copy() 也可以去除 warning，但 .copy() 很少用

df_7 = df[df['age'] > 50].copy()
df_7['preTestScore'] = 7
display(df_7)
display(df)
```

顺便对比一下这两种得到 column index 的方法

```python
df.columns.get_indexer(['preTestScore'])
```

```python
df.columns.get_indexer(['preTestScore', 'female'])
```

```python
df.columns.get_loc('preTestScore')    # 不接受 list，必须只能是一个 string，否则会报错
                                      # doc 上写 get_loc 已经被deprecate了
```

#### 自己的一个发现：

```python
# 这样做会有 warning
df_sub_1 = df[['preTestScore', 'female', 'postTestScore']]
df_sub_1['preTestScore'] = 10
display( df_sub_1 )
display( df )
```

```python
# 但这样居然就没有 warning
df_sub_2 = df.loc[:,['preTestScore', 'female', 'postTestScore']]
df_sub_2['preTestScore'] = 10
display( df_sub_2 )
display( df )

# 注意：这里原 df 并没有被改变
```

```python
# 这样与不会有 warning
df_sub_3 = df.loc[[1, 3],['preTestScore', 'female', 'postTestScore']]
df_sub_3['preTestScore'] = 12
display( df_sub_3 )
display( df )

# 注意：这里原 df 并没有被改变
```
